```python
# Load monthly return series
mvp_ret = pd.read_csv("mvp_monthly_returns.csv")
vw_ret = pd.read_csv("vw_monthly_returns.csv")

mvp_ret["date"] = pd.to_datetime(mvp_ret["date"])
vw_ret["date"] = pd.to_datetime(vw_ret["date"])

# Merge monthly returns
excel_returns = pd.merge(
    vw_ret[["date", "Rp_vw"]],
    mvp_ret[["date", "Rp_mvp"]],
    on="date",
    how="inner"
).sort_values("date").reset_index(drop=True)

# Helper for annualised cumulative return
def annualised_cumulative_return(series):
    vals = series.to_numpy(dtype=float)
    T = len(vals)
    cum_ret = np.prod(1 + vals) - 1
    return (1 + cum_ret) ** (12 / T) - 1

# Summary stats for VW
vw_vals = excel_returns["Rp_vw"]
vw_summary = {
    "ann_avg_return": vw_vals.mean() * 12,
    "ann_volatility": vw_vals.std(ddof=1) * np.sqrt(12),
    "ann_cum_return": annualised_cumulative_return(vw_vals),
    "sharpe_ratio": (vw_vals.mean() / vw_vals.std(ddof=1)) * np.sqrt(12),
    "min_monthly_return": vw_vals.min(),
    "max_monthly_return": vw_vals.max(),
}

# Summary stats for MVP
mvp_vals = excel_returns["Rp_mvp"]
mvp_summary = {
    "ann_avg_return": mvp_vals.mean() * 12,
    "ann_volatility": mvp_vals.std(ddof=1) * np.sqrt(12),
    "ann_cum_return": annualised_cumulative_return(mvp_vals),
    "sharpe_ratio": (mvp_vals.mean() / mvp_vals.std(ddof=1)) * np.sqrt(12),
    "min_monthly_return": mvp_vals.min(),
    "max_monthly_return": mvp_vals.max(),
}

print("VALUE-WEIGHTED")
print(vw_summary)
print("\nMINIMUM-VARIANCE")
print(mvp_summary)

print("\nMonthly returns for Excel:")
print(excel_returns.head())
```
